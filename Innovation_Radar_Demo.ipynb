```python
import pandas as pd
file_path = '/innovation_data.xlsx'
data = pd.read_excel(file_path, skiprows=1)
data.columns = ['Technology', 'Social Acceptance', 'Environmental Benefit', 'Economic Potential', 'Feasibility']
print(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

#  Prepare data
categories = ['Social Acceptance', 'Environmental Benefit', 'Economic Potential', 'Feasibility']
num_vars = len(categories)

#  Compute angles for radar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the loop

#  Create bigger figure
fig, ax = plt.subplots(figsize=(7,7), subplot_kw=dict(polar=True))

#  Plot each technology
colors = ['red', 'blue', 'green']

for i, color in enumerate(colors):
    tech = data.loc[i]
    values = tech[categories].values
    values = np.append(values, values[0])  # close the loop
    ax.plot(angles, values, color=color, linewidth=2, label=tech['Technology'])
    ax.fill(angles, values, color=color, alpha=0.25)

#  Add category labels and title
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title("Innovation Radar - All Technologies", size=5)
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

plt.show()
```

This is a small innovation radar chart. Each line shows one technology we are looking at.

Example technologies: Cultivated Meat, Plant-based Burger, Insect Protein.

Each color represents a different technology.

Social Acceptance → how much people like it

Environmental Benefit → how good it is for the environment

Economic Potential → how much money it could make

Feasibility → how easy it is to make it happen

The bigger the shape, the better the technology scores overall.
For example, the Plant-based Burger scores high in Social Acceptance and Feasibility, while Insect Protein scores high in Environmental Benefit but lower in Social Acceptance.
This chart helps us see all technologies at a glance and decide where to focus research or improvements.”
